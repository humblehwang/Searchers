```python
body = {
    "settings": {
        "number_of_shards": 3,
        "number_of_replicas": 2
    },
    "mappings": {
        "properties": {
            "sid": {
                "type": "integer"
            },
            "name": {
                "type": "text"
            },
            "age": {
                "type": "integer"
            },
            "class": {
                "type": "keyword"
            }
        }
    }
}
```

```python
result = es.indices.get(index='test') #index指定要get哪個index的資訊
print(result)
```

```python
teacher_mapping = {
    "properties": {
        "tid": {
            "type": "integer"
        },
        "name": {
            "type": "text"
        },
        "age": {
            "type": "integer"
        },
        "class": {
            "type": "keyword"
        },
        "salary": {
            "type": "integer"
        }
    }
}
```

```python
es.indices.put_mapping(index='test', body=teacher_mapping)
```

```python
result = es.indices.get(index='test') #index指定要get哪個index的資訊
print(result)
```
